```python
import pymc3 as pm
import numpy as np
import arviz as az

def ex_6(cured, sick,cured_plac,sick_plac):
  occurences_real = np.concatenate((np.zeros(sick), np.ones(cured)))
  occurences_plac = np.concatenate((np.zeros(sick_plac), np.ones(cured_plac)))
  with pm.Model() as model:
    p_real = pm.Uniform("p_real", lower=0, upper=1)
    p_plac = pm.Uniform("p_plac", lower=0, upper=1)

    obs_real = pm.Bernoulli("obs_real", p_real, observed=occurences_real)
    obs_plac = pm.Bernoulli("oobs_placbs", p_plac, observed=occurences_plac)

    p_observed = pm.Deterministic("p_observed", p_real - p_plac)
    idata = pm.sample(2000, tune=2500)
  az.plot_trace(idata, show=True)
```

```python
lst = [[70,30,50,50],[700,300,500,500],[7,3,5,5]]
# lst = [[70,30,50,50]]
for i in lst:
  print("Patient groups:", i)
  ex_6(i[0], i[1], i[2], i[3])
```
